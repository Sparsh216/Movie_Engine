```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tag'] = movies['crew']+movies['keywords']+movies['cast']+movies['genres']
```

```python
new_df=movies[['id','title','tag']]
drop_indices = np.random.choice(new_df.index, 20000, replace=False)
new_df = new_df.drop(drop_indices)
new_df.shape
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors.shape
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted((list(enumerate(similarity[0]))),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted((list(enumerate(distances))),reverse=True,key=lambda x:x[1])[1:10]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
   
```

```python
recommend('Sudden Death')
```
